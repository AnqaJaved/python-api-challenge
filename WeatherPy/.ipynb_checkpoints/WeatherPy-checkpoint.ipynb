## Starter Code to Generate Random Geographic Coordinates and a List of Cities

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Impor the OpenWeatherMap API key
from api_keys import weather_api_key

# Import citipy to determine the cities based on latitude and longitude
from citipy import citipy
```

### Generate the Cities List by Using the `citipy` Library

```python
# Empty list for holding the latitude and longitude combinations
lat_lngs = []

# Empty list for holding the cities names
cities = []

# Range of latitudes and longitudes (latitude ranges from -90 to 90 and longitude ranges from -180 to 180)
lat_range = (-90, 90)  # Latitude ranges from South Pole to North Pole
lng_range = (-180, 180)  # Longitude ranges from West to East

# Create a set of random latitudes and longitudes within the defined range
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)  # Generate random latitudes
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)  # Generate random longitudes
lat_lngs = zip(lats, lngs)  # Pair the latitudes and longitudes together

# Loop through the latitudes and longitudes to identify the nearest city for each coordinate
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name  # Get the nearest city based on the lat/lng coordinates

    # If the city is not already in the list, then add it to the cities list
    if city not in cities:
        cities.append(city)  # Add unique cities to the list

# Print the number of unique cities to confirm that there are enough cities in the list
print(f"Number of cities in the list: {len(cities)}")  # Output the total count of unique cities
```

## Requirement 1: Create Plots to Showcase the Relationship Between Weather Variables and Latitude

### Use the OpenWeatherMap API to retrieve weather data from the cities list generated in the started code

```python
# Set the API base URL
url = "http://api.openweathermap.org/data/2.5/weather?"

# Define an empty list to fetch the weather data for each city
city_data = []

# Print to logger
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters
record_count = 1
set_count = 1


```

```python
# Loop through all the cities in our list to fetch weather data
for i, city in enumerate(cities):

    # Group cities in sets of 50 for logging purposes
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 0

    # Create endpoint URL with each city
    city_url = f"{url}q={city}&appid={weather_api_key}&units=metric"

    # Log the url, record, and set numbers
    print("Processing Record %s of Set %s | %s" % (record_count, set_count, city))

    # Add 1 to the record count
    record_count += 1

    # Run an API request for each of the cities
    try:
        # Parse the JSON and retrieve data
        city_weather = requests.get(city_url).json()

        # Parse out latitude, longitude, max temp, humidity, cloudiness, wind speed, country, and date
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_date = city_weather["dt"]

        # Append the City information into city_data list
        city_data.append({"City": city,
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Date": city_date})

    # If an error is experienced, skip the city
    except:
        print("City not found. Skipping...")
        pass

    # Pause to avoid rate limiting
    time.sleep(1)

# Indicate that Data Loading is complete
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
# Convert the cities weather data into a Pandas DataFrame
city_data_df = pd.DataFrame(city_data)

# Show Record Count
city_data_df.count()
```

## Requirement 2: Compute Linear Regression for Each Relationship


```python
# Import necessary libraries
from scipy.stats import linregress
import matplotlib.pyplot as plt

# Define a function to create Linear Regression plots
def create_linear_regression_plot(x, y, x_label, y_label, title, filename):
    # Perform linear regression
    slope, intercept, r_value, p_value, std_err = linregress(x, y)

    # Create scatter plot
    plt.figure(figsize=(10, 6))  # Set the size of the plot
    plt.scatter(x, y, color='blue', edgecolor='black', alpha=0.7)  # Scatter plot with blue points and black edges

    # Plot the linear regression line
    plt.plot(x, slope * x + intercept, color='red')  # Plot the red regression line

    # Add the graph properties
    plt.title(title, fontsize=16)  # Add a title to the plot
    plt.xlabel(x_label, fontsize=12)  # Add x-axis label
    plt.ylabel(y_label, fontsize=12)  # Add y-axis label
    plt.grid(True)  # Add grid lines for better readability

    # Display the equation and R-squared value on the plot
    plt.text(0.5, 0.1, f"y = {slope:.2f}x + {intercept:.2f}", transform=plt.gca().transAxes, fontsize=12)  # Equation of the line
    plt.text(0.5, 0.05, f"R² = {r_value**2:.2f}", transform=plt.gca().transAxes, fontsize=12)  # R² value

    # Save the figure
    plt.savefig(filename)  # Save the plot as a PNG image in the specified filename

    # Show the plot
    plt.show()  # Display the plot

```

```python
# Latitude vs. Max Temperature
create_linear_regression_plot(
    city_data_df['Lat'], city_data_df['Max Temp'],  # Latitude and Max Temperature
    "Latitude", "Max Temperature (°C)",  # Labels for the axes
    "Latitude vs. Max Temperature",  # Title for the plot
    "output_data/Fig1_regression.png"  # Filename to save the plot
)

# Latitude vs. Humidity
create_linear_regression_plot(
    city_data_df['Lat'], city_data_df['Humidity'],  # Latitude and Humidity
    "Latitude", "Humidity (%)",  # Labels for the axes
    "Latitude vs. Humidity",  # Title for the plot
    "output_data/Fig2_regression.png"  # Filename to save the plot
)

# Latitude vs. Cloudiness
create_linear_regression_plot(
    city_data_df['Lat'], city_data_df['Cloudiness'],  # Latitude and Cloudiness
    "Latitude", "Cloudiness (%)",  # Labels for the axes
    "Latitude vs. Cloudiness",  # Title for the plot
    "output_data/Fig3_regression.png"  # Filename to save the plot
)

# Latitude vs. Wind Speed
create_linear_regression_plot(
    city_data_df['Lat'], city_data_df['Wind Speed'],  # Latitude and Wind Speed
    "Latitude", "Wind Speed (m/s)",  # Labels for the axes
    "Latitude vs. Wind Speed",  # Title for the plot
    "output_data/Fig4_regression.png"  # Filename to save the plot
)
```
