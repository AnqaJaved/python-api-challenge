### Step 1: Create a map that displays a point for every city in the `city_data_df` DataFrame. The size of the point should be the humidity in each city.

```python
import cartopy
import geoviews
import pyproj
```

```python
# Step 1: Load the CSV file created earlier into a Pandas DataFrame
# This DataFrame contains the weather data for each city, including their latitude, longitude, and humidity levels.
city_data_df = pd.read_csv("output_data/cities.csv")

# Step 2: Create an interactive map using the hvplot library
# The map will display a point for every city in the DataFrame.
# The size of the point corresponds to the humidity level in each city, and the color of the points is also based on humidity.

# Create the map using hvplot for an interactive map
city_map = city_data_df.hvplot.points(
    'Lng', 'Lat',  # Longitude and Latitude columns for the map
    color='Humidity',  # Color of the points based on the humidity levels
    size='Humidity',  # Size of the points is also based on the humidity levels
    cmap='coolwarm',  # Color map for the points (coolwarm is just one option)
    alpha=0.7,  # Transparency of the points for better visibility
    title="Cities and their Humidity Levels",  # Title of the map
    geo=True,  # Use geographic projection to plot cities on a map
    tiles='OSM',  # Map style (Open Street Map)
    width=800,  # Set the width of the map
    height=600  # Set the height of the map
)

# Step 3: Display the map
# This step will show the interactive map with the cities plotted based on their geographic coordinates and humidity levels.
city_map

```

### Step 2: Narrow down the `city_data_df` DataFrame to find your ideal weather condition

```python
# Narrow down cities that fit criteria, for example:
# - Max Temp between 20C and 25C
# - Humidity between 30% and 50%
# - Wind Speed less than 10 m/s

ideal_weather_df = city_data_df[
    (city_data_df['Max Temp'] >= 20) & 
    (city_data_df['Max Temp'] <= 25) &
    (city_data_df['Humidity'] >= 30) & 
    (city_data_df['Humidity'] <= 50) &
    (city_data_df['Wind Speed'] <= 10)
]

# Drop any rows with null values
ideal_weather_df = ideal_weather_df.dropna()

# Display sample data
ideal_weather_df.head()
```

### Step 3: Create a new DataFrame called `hotel_df`.

```python
# Use the Pandas copy function to create a DataFrame called hotel_df
# that contains the city, country, coordinates (latitude and longitude), and humidity
hotel_df = ideal_weather_df[['City', 'Country', 'Lat', 'Lng', 'Humidity']].copy()

# Add an empty column called "Hotel Name" to the DataFrame
hotel_df['Hotel Name'] = ""

# Display sample data to verify the new DataFrame
hotel_df.head()
```

### Step 4: For each city, use the Geoapify API to find the first hotel located within 10,000 metres of your coordinates.

```python
# Set the radius to 10,000 meters (10 km)
radius = 10000  # Radius for hotel search in meters

# Set parameters for the API request
params = {
    'apiKey': geoapify_key,  # Use your Geoapify API key
    'categories': 'accommodation.hotel',  # Filter for hotels only
    'limit': 1,  # Limit the results to 1 hotel per search
    'radius': radius  # Set the radius for the search
}

# Print a message to follow up the hotel search
print("Starting hotel search")


# Iterate through the hotel_df DataFrame
for index, row in hotel_df.iterrows():
    # Get latitude and longitude from the DataFrame
    latitude = row['Lat']
    longitude = row['Lng']

    # Add the current city's latitude and longitude to the params dictionary
    params["filter"] = f"circle:{longitude},{latitude},{radius}"  # Set filter with circle radius
    params["bias"] = f"point:{longitude},{latitude}"  # Set bias to the current city coordinates

    # Set the base URL for Geoapify Places API
    base_url = "https://api.geoapify.com/v2/places"

    # Make an API request using the params dictionary
    response = requests.get(base_url, params=params)

    # Convert the API response to JSON format
    name_address = response.json()

    # Grab the first hotel from the results and store the name in the hotel_df DataFrame
    try:
        hotel_df.loc[index, "Hotel Name"] = name_address["features"][0]["properties"]["name"]
    except (KeyError, IndexError):
        # If no hotel is found, set the hotel name as "No hotel found".
        hotel_df.loc[index, "Hotel Name"] = "No hotel found"

    # Log the search results
    print(f"{hotel_df.loc[index, 'City']} - nearest hotel: {hotel_df.loc[index, 'Hotel Name']}")

# Display the updated hotel_df with hotel names
hotel_df
```
